## Data Cleaning

```python
# TO DO
# Dropping unneccesary columns:
#        "sofifa_id","player_url",'club_team_id','league_level','nationality_id','nation_team_id',
#        "work_rate","body_type","player_tags","player_traits","ls",'st','rs',
#        'lw','lf','cf','rf','rw','lam','cam','ram',
#        'lm','lcm','cm','rcm','rm','lwb', 'ldm','cdm','rdm','rwb',
#        'lb','lcb','cb','rcb','rb','gk','player_face_url','club_logo_url','club_flag_url','nation_logo_url',
#        'nation_flag_url'
df = df.drop(labels=["sofifa_id","player_url",'club_team_id','league_level','nationality_id','nation_team_id',
        "work_rate","body_type","player_tags","player_traits","ls",'st','rs',
        'lw','lf','cf','rf','rw','lam','cam','ram',
        'lm','lcm','cm','rcm','rm','lwb', 'ldm','cdm','rdm','rwb',
        'lb','lcb','cb','rcb','rb','gk','player_face_url','club_logo_url','club_flag_url','nation_logo_url',
        'nation_flag_url'], axis=1)
```

```python
# TO DO 
# Split 'player_positions' using comma and only keep the first position
# Hint: You can using string functions OR
# You can split the feature 'player_positions' using comma, keep the first column and drop all others.
# Check our lecture Notebook: feature splitting
df['player_positions'] = df.player_positions.str.split(", ").map(lambda v: v[0])

# DO NOT use for loop like the following:
# for index, value in df['player_positions'].items():  
#     value = value.split(",")[0]
#     df['player_positions'].at[index] = value


# TO DO
# map the position to the reuqired position value:
# 'GK': 'GK', 'CAM': 'CM','CB': 'CB', 'CDM': 'CM',
# 'CM': 'CM','LB': 'LB','LM': 'LM','LW': 'ST',
# 'RB': 'RB','RM': 'RM','RW': 'ST','ST': 'ST',
# 'CF': 'ST','LAM': 'LM','LCB': 'CB','LCM': 'CM',
# 'LDM': 'LM','LF': 'ST','LS': 'ST', 'LWB': 'LB',
# 'RAM': 'RM', 'RCB': 'CB','RCM': 'CM', 'RDM': 'RM',
# 'RF': 'ST', 'RS': 'ST','RWB': 'RB'

def position_mapper(in_pos):
    position = {'GK': 'GK', 'CAM': 'CM','CB': 'CB', 'CDM': 'CM',
    'CM': 'CM','LB': 'LB','LM': 'LM','LW': 'ST',
    'RB': 'RB','RM': 'RM','RW': 'ST','ST': 'ST',
    'CF': 'ST','LAM': 'LM','LCB': 'CB','LCM': 'CM',
    'LDM': 'LM','LF': 'ST','LS': 'ST', 'LWB': 'LB',
    'RAM': 'RM', 'RCB': 'CB','RCM': 'CM', 'RDM': 'RM',
    'RF': 'ST', 'RS': 'ST','RWB': 'RB'}
    return position[in_pos]

df['player_positions'] = df.player_positions.apply(position_mapper)
df.head(10)
```

```python
# Check the missing values of each column
df.info()
```

```python
# TO DO 
# Fill missing values in 'value_eur' with the value at 5% quantile of 'value_eur'
# Fill missing values in 'wage_eur' with the value at 5% quantile of 'wage_eur'
# Fill missing values in 'release_clause_eur' with the mean of 'release_clause_eur'
df['value_eur'] = df['value_eur'].fillna(df['value_eur'].quantile(0.05))
df['wage_eur'] = df['wage_eur'].fillna(df['wage_eur'].quantile(0.05))
df['release_clause_eur'] = df['release_clause_eur'].fillna(df['release_clause_eur'].mean())

# Check if the missing values are filled
df.info()

# BTW
# Remove rows with nation_position = NaN
df = df.dropna(subset=['nation_position'])
df.info()
```

## Feature Encoding

```python
# TO DO
# Use One-Hot encoding to encode 'preferred_foot'
from sklearn.preprocessing import OneHotEncoder 

ohe = OneHotEncoder() 
encoded_data = ohe.fit_transform(df[['preferred_foot']]).toarray()
encoded_data = pd.DataFrame(data = encoded_data, columns=['Left', 'Right'])

df = df.join(encoded_data)

df.head()
```

```python
# TO DO 
# Use ordinal Encoding to encode 'nationality_name'
from sklearn.preprocessing import OrdinalEncoder

categories = list(df.nationality_name.unique().reshape(1, 33))
order_encoder=OrdinalEncoder(categories='auto')
df['encoded_nationality'] = order_encoder.fit_transform(df['nationality_name'].to_numpy().reshape(-1, 1))
df
```

## Feature Scaling

```python
# TO DO
# Use Min-Max Scaler to scale 'skill_moves' and 'weak_foot' to range [0, 100]
from sklearn.preprocessing import MinMaxScaler

skill_moves_scaler = MinMaxScaler(feature_range=(0, 500))
weak_foot_scaler = MinMaxScaler(feature_range=(0, 500))
skill_moves_scaler.fit(df['skill_moves'].values.reshape(-1,1)) 
weak_foot_scaler.fit(df['weak_foot'].values.reshape(-1,1))

df['skill_moves_scaler'] = skill_moves_scaler.transform(df['skill_moves'].values.reshape(-1, 1))
df['weak_foot_scaler'] = weak_foot_scaler.transform(df['weak_foot'].values.reshape(-1, 1))
df
```
