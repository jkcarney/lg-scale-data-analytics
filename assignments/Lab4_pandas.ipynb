## Get Started with pandas

```python
#TO DO
#importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
#TO DO
#read the csv file 'Pokemon.csv' and save it as a pandas dataframe
df = pd.read_csv('Pokemon.csv')
#display the first 10 rows of the table
df.head(10)
```

```python
#TO DO
#change column header into upper case and display the first 5 rows
df = df.rename(columns=str.upper)
df.head(5)
```

```python
#TO DO
#Use mask and index to Show the first 5 legendary pokemons

df.loc[df['STATUS'] == 'Legendary'].head(5)
```

```python
#TO DO
#change and set the index to the name attribute
df = df.set_index('NAME')
df
```

```python
# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

df.head(50)
```

```python
# TO DO
#print the dataframe columns
print(f'The columns of the dataset are: {str(df.columns)}')

#print the shape of the dataframe
print(f'The shape of the dataframe is:  {df.shape}')
```

## Data Cleaning

```python
#TO DO
#drop the column 'SPECIES' (don't forget axis=1)
df = df.drop('SPECIES', axis=1)

# Rename the column 'UNNAMED: 0' to 'NUMBER
df = df.rename(columns={'UNNAMED: 0' : 'NUMBER'})
df
```

```python
#TO DO
#some values in TYPE_2 are empty (NaN) and thus they have to be filled or deleted
#fill NaN values in TYPE_2 with corresponding values of TYPE_1

df.TYPE_2.fillna(df.TYPE_1, inplace=True)
df
```

## Indexing

```python
#TO DO
#retrieve and print complete row data from index with value Bulbasaur (use loc)
df.loc['Bulbasaur']


#retrieve and print complete row date from index 0 (do not use loc)
df.iloc[0]

```

```python
#TO DO
#filtering pokemons using logical operators
# display the pokemons with both Fire type (as type_1) and Dragon type (as type_2)
df[ (df['TYPE_1'] == 'Fire') & (df['TYPE_2'] == 'Dragon') ]
```

## Basic Statistics

```python
#TO DO
#return and print the pokemon (entire row) with highest HP using argmax()
df.iloc[np.argmax(df['HP'])]


#return and print the pokemon with highest DEFENSE using idxmax()
df.loc[df['DEFENSE'].idxmax()]

```

```python
#TO DO
#Find the strongest 3 pokemons (arranges the pokemons in the descendng order of the Total_Points)
#Hint: sort_values() is used for sorting and ascending=False is making it in descending order

df.sort_values(by='TOTAL_POINTS', ascending=False).head(3)
```

```python
#TO DO
#print all the unique types in column TYPE_1
#Hint: use unique()
print(pd.unique(df['TYPE_1']))

#print the count of unique values in column TYPE_1
#Hint: use nunique()
print(df.nunique()['TYPE_1'])
```

```python
#TO DO
#count different types of pokemons (TYPE_1 and TYPE_2 should be counted separately)
#Hint: You can use value_counts()
#Hint: You can also use groupby(attribute).size()
print(df['TYPE_1'].value_counts())
print(df['TYPE_2'].value_counts())


#counts for a single value (count the number of pokemons of bug type as TYPE_1)
print(df.loc[df['TYPE_1'] == 'Bug']['TYPE_1'].count())
```
